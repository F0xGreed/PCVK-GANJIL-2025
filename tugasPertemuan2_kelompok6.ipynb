```python
import cv2 as cv
import matplotlib.pyplot as plt

# 1. Load gambar KTM
img = cv.imread("/content/drive/MyDrive/PCVK GANJIL D4/Gambar/ktmGUWAH.jpg")

# Copy supaya gambar asli tidak berubah
masked = img.copy()

# 2. Tutup tulisan Month/Year (rectangle merah)
cv.rectangle(masked, (820, 580), (990, 610), (0,0,255), -1)

# 3. Tutup tulisan Valid Thru (circle biru)
cv.circle(masked, (770, 650), 40, (255,0,0), -1)   # (x,y), radius, warna

# 4. Tutup angka tanggal (rectangle hijau)
cv.rectangle(masked, (820, 630), (1090, 700), (0,255,0), -1)

# 5. Tampilkan hasil dengan koordinat
plt.figure(figsize=(10,6))
plt.imshow(cv.cvtColor(masked, cv.COLOR_BGR2RGB))
plt.title("Kelompok 6 - Tutup Month/Year, Valid Thru (circle), dan Angka")
plt.grid(True)  # tampilkan grid
plt.xlabel("Koordinat X")
plt.ylabel("Koordinat Y")
plt.show()
```
