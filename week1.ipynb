```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pyplot as plt
```

```python
#list untuk menyimpan url dari beberapa image
urls = ["https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg",
        "https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg",
        "https://image.idntimes.com/post/20190108/maine-coon-cat-photography-robert-sijka-17-57ad8ed924355-880-ce2ed61f54a726dda73489e4653b7bed.jpg?tr=w-750,f-webp,q-75&width=750&format=webp&quality=75"]

#baca dan tampilkan image
#loop pada tiap url image, beberapa image dapat disimpan pada list
for url in urls:
  image = io.imread(url)
  image = cv.resize(image, (0,0), fx=0.5, fy=0.5)
  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
  final_frame = cv.hconcat([image, image_2])
  cv2_imshow(final_frame)
  print('\n')
```

```python
tinggi = image_2.shape[0]
lebar = image_2.shape[1]
print("resolusi image : tinggi x lebar = ", tinggi , " x ", lebar)
cv2_imshow(image_2)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

#membuat garis horizontal di tengah image
for y in range(lebar):
  image_3[int((tinggi)/2), y] = [255, 255, 255]

final_frame = cv.hconcat([image_2, image_3])
cv2_imshow(final_frame)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# garis vertikal di tengah
for x in range(tinggi):
    image_3[x, int(lebar/2)] = [255, 255, 255]

final_frame = cv.hconcat([image_2, image_3])
cv2_imshow(final_frame)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# garis menyilang (X)
for x in range(tinggi):
    y1 = int(x * lebar / tinggi)          # diagonal  proporsional
    y2 = lebar - int(x * lebar / tinggi) - 1  # diagonal  proporsional

    image_3[x, y1] = [255, 255, 255]
    image_3[x, y2] = [255, 255, 255]


final_frame = cv.hconcat([image_2, image_3])
cv2_imshow(final_frame)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# posisi tengah secara vertikal
y_tengah = int(tinggi / 2)

# tentukan panjang garis (misalnya setengah lebar gambar)
panjang = int(lebar / 2)

# mulai dari kolom tengah - setengah panjang, sampai tengah + setengah panjang
start_x = int((lebar - panjang) / 2)
end_x   = start_x + panjang

for y in range(start_x, end_x):
    image_3[y_tengah, y] = [255, 255, 255]

final_frame = cv.hconcat([image_2, image_3])
cv2_imshow(final_frame)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# tentukan posisi kotak (x: vertikal, y: horizontal)
x_awal, y_awal = 100, 150      # titik kiri-atas kotak
tinggi_kotak   = 50
lebar_kotak    = 80

# isi kotak dengan putih
image_3[x_awal:x_awal+tinggi_kotak, y_awal:y_awal+lebar_kotak] = [255, 255, 255]

final_frame = cv.hconcat([image_2, image_3])
cv2_imshow(final_frame)
```
